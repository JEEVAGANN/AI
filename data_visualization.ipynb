```python
## Line plot data
x  = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y1 = [1, 3, 5, 3, 1, 3, 5, 3, 1]
y2 = [2, 4, 6, 4, 2, 4, 6, 4, 2]

```

```python
p.plot(x)
```

```python
p.plot(x,y1)
```

```python
p.xlabel("x")
p.ylabel("Y")
p.plot(x,y1,label='y1',linewidth=10)
p.plot(x,y2,label='y2')
p.title("Multi line chart")
p.legend()
p.show()
```

```python
def random_plots():
  xs = []
  ys = []


  for i in range(20):
    x = i
    y = np.random.randint(10)


    xs.append(x)
    ys.append(y)


  return xs, ys
  random_plots()
```

```python


# Plot subplots
fig = p.figure()


ax1 = p.subplot2grid(shape=(5,2), loc=(0,0), rowspan=1, colspan=2)
ax2 = p.subplot2grid(shape=(5,2), loc=(1,0), rowspan=2, colspan=2)
ax3 = p.subplot2grid(shape=(5,2), loc=(4,0), rowspan=1, colspan=1)
ax4 = p.subplot2grid(shape=(5,2), loc=(4,1), rowspan=1, colspan=1)


x, y = random_plots()
ax1.plot(x, y)


x, y = random_plots()
ax2.plot(x, y)


x, y = random_plots()
ax3.plot(x, y)


x, y = random_plots()
ax4.plot(x, y)
```
