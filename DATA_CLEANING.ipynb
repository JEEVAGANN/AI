```python
#outlier detectio usig 2 score
from scipy import stats
h=np.abs(stats.zscore(data['normalized-losses']))
h
```

```python
h.plot()
```

```python
threshold=h.max()
index_val=np.where(h>=threshold)
data.iloc[index_val]['normalized-losses']
data
```

```python
#drop the outlier value
new_data=data.drop(index=data.iloc[index_val]['normalized-losses'].index)
new_data
```
