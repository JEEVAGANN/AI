```python
import numpy as mp;
import pandas as pd;
import matplotlib.pyplot as plt
```

```python
test=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/test.csv")
train=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/train.csv")
```

```python
train.set_index("PassengerId",inplace=True)
test.set_index("PassengerId",inplace=True)
```

```python
train
```

```python
y_train=train["Survived"]
```

```python
train.drop(labels="Survived",axis=1,inplace=True)
```

```python
train
```

```python
train.shape
```

```python
train1=train.append(test)
```

```python
train1
```

```python
train1.shape
```

```python
colum_drop=["Name","Age","Ticket","Cabin","Parch","Embarked","SibSp"]
train1.drop(labels=colum_drop, axis=1,inplace=True)
```

```python
train1_dummies=pd.get_dummies(train1, columns={"Sex"})
```

```python
train1_dummies.shape
```

```python
train1_dummies.isnull().sum()
```

```python
train1_dummies.fillna(value=0.0, inplace=True)
```

```python
x_train=train1_dummies.values[0:891]
x_test=train1_dummies.values[891:]
```

```python
from sklearn.preprocessing import MinMaxScaler
Scaler=MinMaxScaler()
x_train_Scale=Scaler.fit_transform(x_train)
x_test_Scale=Scaler.fit_transform(x_test)
```

```python
from sklearn.model_selection import train_test_split
x_train_sub,x_validation_sub,y_train_sub,y_validation_sub=train_test_split(x_train_Scale,y_train,random_state=0)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
learning_rates = [0.05, 0.1, 0.25, 0.5, 0.75, 1]
for learning_rate in learning_rates:
    gb = GradientBoostingClassifier(n_estimators=20, learning_rate = learning_rate, max_features=2, max_depth = 2, random_state = 0)
    gb.fit(x_train_sub, y_train_sub)
    print("Learning rate: ", learning_rate)
    print("Accuracy score (training): {0:.3f}".format(gb.score(x_train_sub, y_train_sub)))
    print("Accuracy score (validation): {0:.3f}".format(gb.score(x_validation_sub, y_validation_sub)))
    print()
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
learning_rates = [0.05, 0.1, 0.25, 0.5, 0.75, 1]
for learning_rate in learning_rates:
    gb = AdaBoostClassifier(n_estimators=20, learning_rate = learning_rate, random_state = 0)
    gb.fit(x_train_sub, y_train_sub)
    print("Learning rate: ", learning_rate)
    print("Accuracy score (training): {0:.3f}".format(gb.score(x_train_sub, y_train_sub)))
    print("Accuracy score (validation): {0:.3f}".format(gb.score(x_validation_sub, y_validation_sub)))
    print()
```
